# Data Cleaning & Feature Engineering

## Objective
Clean the raw Titanic dataset and prepare it for machine learning models.
This includes handling missing values, encoding categorical variables,
and creating useful features.


```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/kaggle/input/data-train2/train.csv")
data.head()
```

## Step 1: Drop irrelevant columns

```python
data=data.drop(columns=["Ticket"])
data=data.drop(columns=["Name"])
data=data.drop(columns=["PassengerId"])
data=data.drop(columns=["Cabin"])#Cabin has too many missing values, so it will be dropped.

```

## Step 2: Handle missing values

### Age
Age has missing values. We fill them using the median to reduce
the effect of outliers.

```python
data["Age"]=data["Age"].fillna(data["Age"].median())

```

Embarked has a few missing values. We fill them using the most common value.

```python
data["Embarked"] = data["Embarked"].fillna(data["Embarked"].mode()[0])
```

## Step 3: Encode categorical variables

Convert categorical features into numerical values for ML models.


```python
data["Sex"]=data["Sex"].map({"male": 0 ,"female":1})
data=pd.get_dummies(data,columns=["Embarked"],drop_first=True)
```

## Step 4: Feature engineering

Create new features that may improve model performance.


```python
data["FamilySize"]=data["Parch"]+data["SibSp"]+1
```
